4. correct the inconsistent data

```python
display(df.head(10))
```

```python
#  checking the unique value in each columns
print("Number of {Transaction ID} Unique: ",df['Transaction ID'].nunique())
print("Item Unique: ",df['Item'].unique())
print("Quantity Unique: ",df['Quantity'].unique())
print("Price Per Unit Unique: ",df['Price Per Unit'].unique())
print("Total Spent Unique: ",df['Total Spent'].unique())
print("Payment Method Unique: ",df['Payment Method'].unique())
print("Location Unique: ",df['Location'].unique())
```

--> delete the row that Quantity is NaN or missing value(not digits) and one of the columns( Price Per Unit, Total Spent ) is NaN or Missing value because we won't be able to find the Quantity 

```python
bad_rows = df[(~df["Quantity"].astype(str).str.isdigit()) & ((df['Price Per Unit'].isnull()) | (df['Total Spent'].isnull()))]
df = df.drop(bad_rows.index)
display(df[(~df["Quantity"].astype(str).str.isdigit()) & ((df['Price Per Unit'].isnull()) | (df['Total Spent'].isnull()))])
```

```python
# changing the Quantity base on the value fro Price Per Unit and Total Spent
Quantity_not_number = ~df["Quantity"].astype(str).str.isdigit()
df.loc[Quantity_not_number, "Quantity"] = 0
```

```python
df_check = df[(df["Quantity"] == 0) & (~df['Price Per Unit'].isnull()) & (~df['Total Spent'].isnull())]
display(df_check)
```

```python
# change data type of Quantity to int64 and find the Quantity by Price Per Unit * Total Spent
df['Quantity'] = pd.to_numeric(df['Quantity'])
df_check = df[(df["Quantity"] == 0) & (~df['Price Per Unit'].isnull()) & (~df['Total Spent'].isnull())]
df.loc[df_check.index, "Quantity"] = (
    pd.to_numeric(df.loc[df_check.index, "Total Spent"], errors="coerce") /
    pd.to_numeric(df.loc[df_check.index, "Price Per Unit"], errors="coerce"))
display(df[df['Quantity'].isnull()])
```

```python
# delete the row that we can't find the Quantity Quantity 
Q_bad_row = df[df['Quantity'].isnull()]
df = df.drop(Q_bad_row.index)
print(df['Quantity'].isnull().any())
```

```python
display(df.head())
```

```python
# --> checking on Price Per Unit
df[((df["Price Per Unit"] == 'ERROR') | (df["Price Per Unit"] == 'UNKNOWN')| (df["Price Per Unit"].isnull()))
& (df["Total Spent"] != 'ERROR') & (df["Total Spent"] != 'UNKNOWN') & (~df["Total Spent"].isnull())]
```

```python
# finding the value of the missing value in Price Per Unit = Total Spent/Quantity
df_PricePerUnit = df[((df["Price Per Unit"] == 'ERROR') | (df["Price Per Unit"] == 'UNKNOWN')| (df["Price Per Unit"].isnull()))
& (df["Total Spent"] != 'ERROR') & (df["Total Spent"] != 'UNKNOWN') & (~df["Total Spent"].isnull())]
df.loc[df_PricePerUnit.index, "Price Per Unit"] = (
    pd.to_numeric(df.loc[df_PricePerUnit.index, "Total Spent"], errors="coerce") /
    pd.to_numeric(df.loc[df_PricePerUnit.index, "Quantity"], errors="coerce"))
```

```python
# delete the rest of the row that Price Per Unit is still NaN
df = df.drop(df[df['Price Per Unit'].isnull()].index)
print(df['Price Per Unit'].isnull().any())
```

```python
# check on Total Spent
df[(df['Total Spent'].isnull()) | (df['Total Spent'] == 'ERROR') | (df['Total Spent'] == 'UNKNOWN')]
```

```python
# finding the Total Spent = Quantity * Price Per Unit
df_TotalSpent = df[(df['Total Spent'].isnull()) | (df['Total Spent'] == 'ERROR') | (df['Total Spent'] == 'UNKNOWN')]
df.loc[df_TotalSpent.index, "Total Spent"] = (
    pd.to_numeric(df.loc[df_TotalSpent.index, "Quantity"], errors="coerce") *
    pd.to_numeric(df.loc[df_TotalSpent.index, "Price Per Unit"], errors="coerce"))
```

```python
# still NaN cause one columns( Price Per Unit is NaN) 
df[(df['Total Spent'].isnull()) | (df['Total Spent'] == 'ERROR') | (df['Total Spent'] == 'UNKNOWN')]
```

```python
# delete the row if we can't find Total Spent
df = df.drop(df[(df['Total Spent'].isnull()) | (df['Total Spent'] == 'ERROR') | (df['Total Spent'] == 'UNKNOWN')].index)
print(df['Total Spent'].isnull().any())
```

```python
# change the data type of those 3 colums
df["Quantity"] = df['Quantity'].astype(int)
df["Price Per Unit"] = pd.to_numeric(df["Price Per Unit"], errors="coerce")
df["Total Spent"] = pd.to_numeric(df["Total Spent"], errors="coerce")
df.info()
```

```python
display(df[(df['Item'].isnull()) | (df['Item'] == 'UNKNOWN') | (df['Item'] == 'ERROR')].sort_values('Transaction ID'))
```

```python
df["Item"] = df["Item"].replace(["UNKNOWN", "ERROR"], "UNKNOWN_ITEM")
df["Item"] = df["Item"].fillna("UNKNOWN_ITEM")
display(df[(df['Item'].isnull()) | (df['Item'] == 'UNKNOWN') | (df['Item'] == 'ERROR')])
```

```python
df.info()
```

```python
# Replace 'UNKNOWN' and 'ERROR' with NaN for consistency
df['Location'] = df['Location'].replace(['UNKNOWN', 'ERROR'], np.nan)
df['Payment Method'] = df['Payment Method'].replace(['UNKNOWN', 'ERROR'], np.nan)
```

```python
# Impute missing values with the mode
modeLocation = df['Location'].mode()[0]
modePaymentMethod = df['Payment Method'].mode()[0]

df['Location'].fillna(modeLocation)
df['Payment Method'].fillna(modePaymentMethod)
```

```python
df.info()
```

```python
df[df['Transaction Date'].isnull()]
```

```python
# Convert column to datetime (errors='coerce' will turn invalid dates into NaN)
df['Transaction Date'] = pd.to_datetime(df['Transaction Date'], errors='coerce')

# Fill NaN with forward fill
df['Transaction Date'] = df['Transaction Date'].ffill()
print(df['Transaction Date'].isnull().any())
```
